```python
# Import the pandas library
import pandas as pd

# Read the CSV file 'zomato.csv' into a DataFrame
df = pd.read_csv('./zomato.csv')

# Drop the 'address' and 'phone' columns from the DataFrame
df.drop(['address', 'phone'], axis=1, inplace=True)

#--- Inspect data ---
df
```

```python
# Rename the 'rate' column to 'rating', 'approx_cost(for two people)' to 'approx_cost', and 'listed_in(type)' to 'type'
df.rename(columns={'rate': 'rating', 'approx_cost(for two people)': 'approx_cost', 'listed_in(type)': 'type'}, inplace=True)

#--- Inspect data ---
df
```

```python
# deleting null values of name column
df.dropna(subset=['name'], inplace=True)

# handling null values of online_order
df['online_order'].fillna('NA', inplace=True)

# changing null values of book_table
df['book_table'].fillna('NA', inplace=True)

# changing null values of rating to zero as it is a numerical datatype
df['rating'].fillna(0, inplace = True)

# changing null values of votes to zero as it is a numerical datatype
df['votes'].fillna(0, inplace=True)

# changing null values of location to NA
df['location'].fillna('NA', inplace=True)

# changing null values of rest_type to NA
df['rest_type'].fillna('NA', inplace=True)

# changing null values of dishliked to NA
df['dish_liked'].fillna('NA', inplace=True)

# changing null values of cuisines to NA
df['cuisines'].fillna('NA', inplace=True)

# changing null values of approxcost to 0 as it is a numerical value
df['approx_cost'].fillna(0, inplace=True)

# changing null values of type to NA
df['type'].fillna('NA', inplace=True)

#--- Inspect data ---
df
```

```python
# droping the duplicates value keeping the first
df.drop_duplicates(keep='first', inplace=True)
#--- Inspect data ---
df
```

```python
# Define the list of columns to check for containing 'RATED' or 'Rated'
columns_to_check = df.columns.tolist()

# Iterate through each column in the DataFrame
for column in columns_to_check:
    # Filter out rows where the current column does not contain 'RATED' or 'Rated', case insensitive
    df = df[df[column].str.contains('RATED|Rated', case=False) == False]
    
#--- Inspect data ---
df
```

```python
# online order table should have only yes and no, remove other values
df = df[df['online_order'].str.contains('Yes|No') == True ]
# check for rating table and replace NEW,- values to 0 and remove /5
df['rating'].replace(to_replace='NEW', value=0, inplace=True)
df['rating'].replace(to_replace='-', value=0, inplace=True)
df['rating'].replace(regex=r'/5', value='', inplace=True)
#--- Inspect data ---
df
```
